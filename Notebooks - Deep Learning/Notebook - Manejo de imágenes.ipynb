# Configuración inicial

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
im = io.imread('https://github.com/jorge-robledo11/Datasets/blob/main/Im%C3%A1genes/Mara.jpeg?raw=true')
```

```python
im.shape
```

```python
plt.imshow(im);
```

```python
r = im[:,:,0]
g = im[:,:,1]
b = im[:,:,2]
```

```python
plt.imshow(r.T,cmap='gray');
```

```python
r.shape

# Capa auxiliar con solo valores en 0
aux_dim = np.zeros([800,600])
```

```python
# Stackeamos la capa para que sea (255,0,0) en RGB
red = np.dstack((r, aux_dim, aux_dim)).astype(np.uint8)
plt.imshow(red);
```

```python
# Stackeamos la capa para que sea (0,255,0) en RGB
green = np.dstack((aux_dim, g, aux_dim)).astype(np.uint8)
plt.imshow(green);
```

```python
# Stackeamos la capa para que sea (0,0,255) en RGB
blue = np.dstack((aux_dim,aux_dim, b)).astype(np.uint8)
plt.imshow(blue);
```
