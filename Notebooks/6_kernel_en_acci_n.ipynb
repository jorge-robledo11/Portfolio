```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
import scipy.ndimage as nd
```

```python
# Cargamos la imagen
img = io.imread('https://github.com/jorge-robledo11/Datasets/blob/main/Im%C3%A1genes/26733338_1814114721993254_5771064522826693294_n.jpg?raw=true')
plt.imshow(img);
```

```python
"""Transformar la imagen a escala de grises, siempre que no importen los colores"""
img_gray = color.rgb2gray(img)
plt.imshow(img_gray);
```

```python
# Kernels (Matrices 3x3)

kernel = np.array([[-1,0,1],
                  [-1,0,1],
                  [-1,0,1]])

kernel2 = np.array([[0,0,0],
                  [0,1,0],
                  [0,0,0]])

kernel3 = np.array([[1,1,1],
                  [1,1,1],
                  [1,1,1]])

kernel4 = np.array([[0,-1,0],
                  [-1,5,-1],
                  [0,-1,0]])
```

```python
# Convolutar una imagen
img_new = nd.convolve(img_gray, kernel)
plt.imshow(img_new);
```

```python
fig, axes = plt.subplots(1,2, figsize=(15,10))

axes[0].imshow(img_gray, cmap=plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(img_new, cmap=plt.cm.gray)
axes[1].set_title('Convolucionada')
axes[1].axis('off')
```

```python
# Convolutar una imagen
img_new2 = nd.convolve(img_gray, kernel2)
plt.imshow(img_new2);
```

```python
fig, axes = plt.subplots(1,2, figsize=(15,10))

axes[0].imshow(img_gray, cmap=plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(img_new2, cmap=plt.cm.gray)
axes[1].set_title('Convolucionada')
axes[1].axis('off')
```
